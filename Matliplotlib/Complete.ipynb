```python
def iterrow_example(df,col):
    for i , row in df.iterrows():
        i =1
        val = row[col] # we store the value in val variable
        if val == 0:
            df.at[i,col] = 1 #we here .at method on data frame at index and colunm[i,col]
       # else:
        #    df.at[i,col] = 0 #we here .at method on data frame at index and colunm[i,col]
iterrow_example(df,"Rank")
df.head(23)
```

```python
%timeit iterrow_example(df,"Rank")
```

```python
def iloc_example(df,col):
    for i in df.index:
        i=1
        val = df[col].iloc[i]
        if val < 1:
            df.at[i,col]=1
        #else:
           # df.at[i,col]=1
iloc_example(df,'Rank')
df.head(3)
```

```python
%timeit iloc_example(df,'Rank')
```

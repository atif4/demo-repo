```python
array = [np.array(['bar','bar','bar','baz','baz','baz','foo','foo','foo','qux','qux','qux']),
        np.array(['height','weight','girth','height','weight','girth','height','weight','girth','height','weight','girth'])
        ]
array
```

```python
multi = pd.DataFrame(np.random.randn(12,4),index=array,columns=['a','b','c','d'])
multi.index.names=['lel0','lvl1']
```

```python
multi
```

```python
multi.drop('girth',level='lvl1',axis=0)# single row
```

```python
multi.drop(['baz','foo'],level=0,axis=0,inplace=True) # multiple rows 
multi
```

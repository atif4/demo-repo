```python
from collections import Counter 
```

```python
def get_tokens(docx,nums=30):
    word_tokens = Counter(docx)
    most_common = word_tokens.most_common(nums)
    result = dict(most_common)
    return result
```

```python
get_tokens(cleaning_user_name_token)
```

```python
def get_most_common (token):
    most_common = get_tokens(token)
    user_info_df = pd.DataFrame(most_common.items(),columns=['words','score'])
    return user_info_df
```
